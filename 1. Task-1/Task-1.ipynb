### **Simple Linear Regression**
In this regression, we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

```python
# Importing necessary Libraries to use
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
```

```python
student_data = pd.read_csv('http://bit.ly/w-data')
student_data.head(10)
```

## **Finding any relation between two data by ploting the student_data**


```python
student_data.plot(x='Hours', y='Scores', style='ro')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

By ploting data in 2D graph, we can clearly see that there is a positive linear relation or correlation between the number of hours studied and percentage of score. 

```python
student_data.corr()
```

The positive linear relation between the number of hours studied and percentage of score is **0.976191** 

## **Preparing the data**

```python
#dividing the data into "attributes" (inputs) and "labels" (outputs)
x = student_data.iloc[0:, :-1].values
y = student_data.iloc[:, 1].values
```

```python
# spliting this data into training and test sets.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### **Training the Algorithm**
We have split our data into training and testing sets, and now is finally the time to train our algorithm. 

```python
regressor = LinearRegression().fit(x_train, y_train)

print("Accuracy of Training Set: {:0.2f}".format(regressor.score(x_train, y_train)))

print("Accuracy of Test Set: {:0.2f}".format(regressor.score(x_test, y_test)))

print("Training is complete")
```

## OLS Model

```python
# Plotting the regression line
line = regressor.coef_*x+regressor.intercept_

# Plotting for the test data
plt.scatter(x, y, c='red')
plt.plot(x, line, c='g');
plt.xlabel("Hours")
plt.ylabel("Percentage")
plt.show()
```

## **Making Prediction**
Now that we have trained our algorithm, it's time to make some predictions.

```python
 # Testing data - In Hours
 print(x_test)

 prediction = regressor.predict(x_test)
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction})  
df.head(10) 
```

```python
hour = 9.25
hours=np.array([9.25])
my_pred = regressor.predict(hours.reshape(-1,1))
print("No of Hours = {}".format(hour))
print("Predicted Score = {}".format(my_pred[0]))
```

** From the above prediction, I have observed that predicted score of a student if he/she studies for 9.25 hours will be around 92~93**

## **Evaluating the model**
The model is working well but it is particularly important to compare how well different algorithms perform on a particular dataset.

**For simplicity here, we have chosen the mean square error. There are many such metrics.**

```python
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, prediction)) 
```
